AI 에이전트가 다양한 실제 환경에서 안정적으로 동작하기 위해서는  
예기치 못한 상황, 오류, 그리고 오작동을 스스로 관리할 수 있어야 한다.  


인간이 예상치 못한 장애물을 만나도 적응하고 해결하듯,  
지능형 에이전트도 문제를 감지하고 복구 절차를 수행하거나  
최소한 통제된 방식으로 실패할 수 있는 견고한 시스템이 필요하다.

이 필수적인 요구 사항이 바로 **Exception Handling and Recovery 패턴의 기반**이 된다.

이 패턴은 다양한 어려움과 비정상 상황에서도  
중단되지 않고 기능을 유지할 수 있는 **내구성 있고 회복력 있는 에이전트**를 만드는 데 초점을 둔다.

아래의 두 가지 전략이 중요한데,

### 1. **사전적(proactive) 준비**
문제가 발생하기 전에 대비하고 위험을 완화하는 전략

### 2. **사후적(reactive) 대응**
문제가 발생했을 때 신속하게 진단·복구하여 운영을 지속하는 전략

이 두 전략을 모두 갖추는 것은 복잡하고 예측 불가능한 환경 속에서도  
에이전트가 성공적으로 기능하는 데 필수적이며,  
에이전트의 **효과성과 신뢰성**을 더욱 높여준다.

예상치 못한 사건을 다루는 능력은  
AI 시스템이 “단지 똑똑한 것”을 넘어서  
**안정적이고 신뢰할 수 있는 시스템**이 되도록 만든다.

--- 

안정성은 아래 두 요소를 통해 더욱 강화된다

- **모니터링 도구**  
- **진단 시스템**  

문제가 발생했을 때 빠르게 인식하고 해결할 수 있게 도와주며,  
잠재적인 중단을 예방하고 지속적인 운영을 가능하게 한다.

**Reflection** 기법과 결합되어 사용되기도 한다.

예시: **self-repair** 루프
- 최초 시도가 실패하고 예외가 발생한다.
- Reflection으로 실패를 분석한다.
- 분석 결과를 기반으로 더 나은 접근(개선된 프롬프트 등)을 설계한다.
- 재시도해서 문제를 해결한다.

# Exception Handling and Recovery 패턴 개요

**Exception Handling and Recovery 패턴**은 AI 에이전트가 운영 과정에서 발생할 수 있는 다양한 실패를 안정적으로 처리할 수 있도록 하는 핵심 설계 방식이다. 이 패턴은 *도구 오류*, *API 지연*, *서비스 불가 상태*, *예상치 못한 예외 상황* 등 잠재적인 문제를 사전에 예측하고, 이러한 문제를 완화하기 위한 전략을 구축하는 데 중점을 둔다.

주요 전략에는 오류 로깅, 재시도(retry), 대체 경로 사용(fallback), 점진적 기능 축소(graceful degradation), 알림(notification)이 포함된다. 또한 복구(recovery)를 위한 핵심 메커니즘으로 상태 롤백(state rollback), 진단(diagnosis), 자기 수정(self-correction), 에스컬레이션(escalation) 등을 활용해 에이전트를 안정적인 상태로 되돌린다.

이 패턴을 적용하면 AI 에이전트는 예측 불가능한 환경에서도 더 높은 신뢰성과 견고함을 확보할 수 있다. 실제 활용 예로는 다음이 있다:

- 데이터베이스 오류를 처리하는 **챗봇**
- 금융 오류에 대응하는 **트레이딩 봇**
- 장치 오작동에 대응하는 **스마트 홈 에이전트**

이 패턴은 복잡성과 실패가 존재하는 환경에서도 에이전트가 **지속적으로 기능을 유지하도록 보장**한다.


---

---

### 오류 감지 (Error Detection)

오류 감지는 운영 중 발생하는 문제를 정확하게 식별하는 단계이다. 예시는 다음과 같다:

- 잘못되거나 비정상적인 도구 출력
- API 오류(404 Not Found, 500 Internal Server Error 등)
- 비정상적으로 긴 응답 시간
- 기대한 형식에서 벗어난 비논리적 출력
- 다른 에이전트 또는 모니터링 시스템에 의한 이상 탐지

이러한 감지 단계는 더 큰 문제로 확대되기 전에 조기 대응할 수 있게 해 준다.

---

### 오류 처리 (Error Handling)

오류가 감지되면, 체계적인 대응 절차가 필요하다. 주요 전략은 다음과 같다:

- **Logging**: 나중에 분석할 수 있도록 오류 정보를 상세하게 기록  
- **Retries**: 일시적 오류일 가능성을 고려하여 약간 수정된 조건으로 재시도  
- **Fallbacks**: 동일 작업을 다른 방식으로 수행하여 핵심 기능 유지  
- **Graceful Degradation**: 완전한 기능 복구가 어렵다면 일부 기능만 유지  
- **Notification**: 사람이 개입해야 하는 상황이라면 운영자 혹은 다른 에이전트에게 알림  

---

### 복구 (Recovery)

복구 단계는 시스템을 안정적인 상태로 되돌리는 과정이다:

- **State Rollback**: 오류 발생 직전의 상태로 되돌려 부작용 제거  
- **Diagnosis**: 오류 원인을 분석하여 재발 방지  
- **Self-Correction**: 에이전트 계획·논리·파라미터를 재조정 또는 재계획  
- **Escalation**: 자동 복구가 불가능하면 인간 혹은 상위 시스템으로 문제 전달  

이 과정을 통해 에이전트는 단순히 오류를 “견디는” 수준을 넘어, **주도적으로 복구하고 더 나아진 상태로 운영**될 수 있게 된다.

---

Exception Handling and Recovery 패턴을 구현하면 AI 에이전트는 다음과 같은 특성을 갖는다:

- 오류 발생 시 즉각적 탐지 및 처리
- 자동 복구 또는 안전한 실패 처리
- 최소한의 다운타임 유지
- 예측 불가능한 환경에서도 안정적 운영
- 사용자에게 더 신뢰도 높은 경험 제공

이 패턴은 취약하고 예측 불가한 시스템을 **강력하고 신뢰 가능한 에이전트**로 변모시키는 핵심 요소이며, 실제 세계에서 AI 시스템을 배포할 때 반드시 고려해야 하는 설계 패턴이다.

```python
from __future__ import annotations
import os
from dotenv import load_dotenv
from typing import TypedDict, Optional

from langgraph.graph import StateGraph, START, END
from langgraph.checkpoint.memory import InMemorySaver
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
```

```python
load_dotenv()

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
GEMINI_MODEL = "gemini-2.5-flash"

llm = ChatGoogleGenerativeAI(
    api_key=GEMINI_API_KEY,
    model=GEMINI_MODEL,
    temperature=0.3
)
```

상태 정의 (TypedDict)

```python
class LocationState(TypedDict, total=False):
    """위치 조회 에이전트의 공용 상태.

    필드:
        user_query: 사용자가 입력한 원본 질의 (주소/질문 등)
        primary_location_failed: 1차 정밀 위치 조회 실패 여부
        location_result: 최종 위치 정보(정밀/범위 정보 포함)
        final_response: 사용자에게 보여줄 최종 자연어 응답
    """
    user_query: str
    primary_location_failed: bool
    location_result: Optional[str]
    final_response: str
```

Tool 함수들

```python
def get_precise_location_info(address: str) -> str:
     """정확한 위치 정보를 가져오는 1차 도구 (예시)."""
     if "fail_precise" in address: # 일부러 실패
         raise RuntimeError("정밀 위치 조회 실패")

     return f"정확한 위치 정보 for {address}"

def get_general_location_info(city: str) -> str:
    """일반 위치 정보를 가져오는 2차 도구 (예시)."""
    if not city:
        raise RuntimeError("도시 정보를 찾을 수 없음")

    return f"{city}의 일반 위치 정보"
```

노드 1: primary_handler (정밀 위치 조회 시도)

```python
def primary_handler_node(state: LocationState) -> dict:
    """정밀 위치 조회를 1차로 시도하는 노드.

    - 성공 시: location_result 저장, primary_location_failed = False
    - 실패 시: primary_location_failed = True 로 설정
    """
    query = state.get("user_query", "")
    try:
        result = get_precise_location_info(query)
        return {
            "location_result": result,
            "primary_location_failed": False
        }
    except Exception:
        return {
            "primary_location_failed": True
        }
```

노드 2: fallback_handler (정밀 조회 실패 시 폴백 수행)

```python
def fallback_handler_node(state: LocationState) -> dict:
    """정밀 조회가 실패했을 경우, 도시 단위 폴백 조회를 수행하는 노드.

    - state["primary_location_failed"]가 True일 때만 동작
    - 사용자의 원본 질의에서 LLM을 이용해 '도시명'만 추출
    - get_general_area_info(city) 도구 호출
    """
    if not state.get("primary_location_failed"):
        # 정밀 조회가 성공했으므로 폴백 불필요
        return {}

    user_query = state.get("user_query", "")

    # LLM에게 도시명만 추출하도록 요청
    prompt = (
        "다음 문장에서 도시명(시/도시)을 한 단어로만 추출하세요.\n"
        "설명이나 다른 단어는 쓰지 말고, 도시명만 출력하세요.\n\n"
        f"문장: {user_query}"
    )
    resp = llm.invoke([HumanMessage(content=prompt)])
    city = str(resp.content).strip()

    try:
        general_info = get_general_location_info(city)
        return {
            "location_result": general_info
        }
    except Exception:
        # 폴백도 실패하면 최종 응답 단계에서 "찾지 못했다"로 처리
        return {}

```

노드 3: response_agent (최종 응답 생성)

```python
def response_agent_node(state: LocationState) -> dict:
    """location_result를 읽어 사용자에게 보여줄 최종 응답을 생성하는 노드."""
    location_result = state.get("location_result")

    if location_result:
        final_response = (
            f"요청하신 위치 정보는 다음과 같습니다:\n{location_result}"
        )
    else:
        final_response = (
            "죄송합니다. 요청하신 위치 정보를 찾지 못했습니다."
        )

    return {"final_response": final_response}
```

```python
builder = StateGraph(LocationState)

builder.add_node("primary_handler", primary_handler_node)
builder.add_node("fallback_handler", fallback_handler_node)
builder.add_node("response_agent", response_agent_node)

builder.add_edge(START, "primary_handler")
builder.add_edge("primary_handler", "fallback_handler")
builder.add_edge("fallback_handler", "response_agent")
builder.add_edge("response_agent", END)

app = builder.compile()
```

```python
# 정밀 조회가 성공하는 예
initial_state_ok: LocationState = {
    "user_query": "서울특별시 강남구 테헤란로 123",
}
config = {"configurable": {"thread_id": "location-thread-1"}}
final_state_ok = graph.invoke(initial_state_ok, config=config)
print("=== 성공 케이스 ===")
print(final_state_ok["final_response"])
```

```python
# 정밀 조회가 실패하고 폴백으로 일반 지역 정보를 가져오는 예
initial_state_fail: LocationState = {
    "user_query": "fail_precise 부산 해운대구 해운대로 456",
}
final_state_fail = graph.invoke(initial_state_fail, config=config)
print("\n=== 폴백 케이스 ===")
print(final_state_fail["final_response"])
```

# At a Glance

## What

운영되는 AI 에이전트는 필연적으로 **예기치 못한 상황, 오류, 시스템 오작동**을 겪게 된다.  
이러한 문제는 다음과 같이 다양하게 발생할 수 있다:

- 도구(tool) 실패  
- 네트워크 연결 문제  
- 잘못된 입력 데이터  
- 외부 서비스 장애  

이러한 장애들은 에이전트의 작업 수행 능력을 위협하며,  
적절한 대응 구조가 없다면 에이전트는 **불안정하고 신뢰할 수 없으며, 예상치 못한 문제로 쉽게 완전 종료될 위험**이 있다.  

---

## 🧠 Why (왜 필요한가?)

이러한 문제를 해결하기 위한 **표준화된 안정성 설계 방식**이 **Exception Handling and Recovery 패턴**이다.
이 패턴은 에이전트에게 다음과 같은 능력을 부여한다:

### 사전적 대응(Proactive)
- 도구 출력, API 응답 등을 모니터링하여  
  장애를 빠르게 감지하고 조기 대응

### 사후적 대응(Reactive)
- 문제 상황에 대한 상세 로깅(진단용)
- 일시적 오류에 대한 재시도(retry)
- 대체 전략(fallback) 실행

### 심각한 오류 처리를 위한 복구 전략(Recovery Protocols)
- 상태 롤백
- self-correction
- 사람에게 에스컬레이션

구조적 접근으로
**운영 무결성(operational integrity)**을 유지하고,  
실패에서 학습하며, 예측 불가능한 환경에서도 **안정적으로 동작**할 수 있다.

# 결론 (Conclusion)

이 장에서는 **Exception Handling and Recovery 패턴**을 다루며,  
견고하고 신뢰할 수 있는 AI 에이전트를 개발하기 위해 이 패턴이 왜 중요한지 설명한다.

이 패턴은 AI 에이전트가:

- 예상치 못한 문제를 **인식하고**,  
- 적절하게 **대응하며**,  
- 안정적인 운영 상태로 **복귀**할 수 있도록 설계된 구조이다.

여기에는 다음 요소들이 포함된다:

- 오류 감지(error detection)  
- 로깅, 재시도, 폴백(fallback) 등의 처리 메커니즘  
- 상태 복구(state recovery) 전략  

또한 다양한 실제 응용 사례(고객 서비스, 금융 거래, 스마트 홈, 로봇 공정 등)를 통해  
이 패턴이 **현실 세계의 복잡성과 실패 상황**을 다루는 데 얼마나 중요한지 보여준다.

결과적으로 Exception Handling and Recovery 패턴은  
AI 에이전트가 **신뢰성, 적응성, 안정성**을 갖추고  
동적인 환경에서도 꾸준한 성능을 제공하도록 돕는 핵심 설계 요소이다.
