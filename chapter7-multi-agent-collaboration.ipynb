# Chapter 7: Multi-Agent Collaboration

### 1. 단일형(monolithic) 에이전트의 한계
단일 에이전트 구조는 명확히 정의된 문제에는 효과적이지만, 복잡한 Multi-domain 문제를 다루기에는 한계가 있다. 이러한 구조에서는 하나의 에이전트가 모든 기능을 담당하기 때문에, 유연성과 확장성이 제한된다.

### 2. 다중 에이전트 협업(Multi-Agent Collaboration) 개념
이 한계를 극복하기 위해 **멀티 에이전트 협업 패턴**이 제안된다. 멀티 에이전트 협업 패턴은 시스템을 여러 **전문화된 에이전트들의 협력적 집합체(cooperative ensemble)** 로 구성하는 접근 방식이다.  
핵심 원리는 **작업 분해(task decomposition)** 로, 고수준 목표를 여러 하위 과제로 나누고, 각 과제는 해당 작업에 최적화된 능력(도구, 데이터 접근권, 추론 능력 등)을 가진 에이전트에게 할당된다.

### 3. 예시
예를 들어, 복잡한 연구 질의는 다음과 같이 분해될 수 있다.
- **Research Agent**: 정보 검색 담당  
- **Data Analysis Agent**: 통계 분석 수행  
- **Synthesis Agent**: 결과 보고서 작성  

이 협업의 핵심은 단순한 역할 분담이 아니라, **에이전트 간의 효율적인 통신 메커니즘**이다.

### 4. 통신과 조정
다중 에이전트 간 협력을 위해서는 다음 요소들이 필요하다.
- **표준화된 통신 프로토콜**  
- **공유된 온톨로지(shared ontology)**  

이를 통해 에이전트들은 데이터를 교환하고, 하위 작업을 위임하며, 행동을 조정하여 일관된 최종 결과를 생성할 수 있다.

### 5. 장점
이러한 분산형 아키텍처는 다음과 같은 장점을 제공한다.
- **모듈성(Modularity)**: 각 에이전트가 독립적으로 개발·관리 가능  
- **확장성(Scalability)**: 새로운 기능을 쉽게 추가 가능  
- **강건성(Robustness)**: 일부 에이전트가 실패해도 전체 시스템이 중단되지 않음  

결과적으로, 다중 에이전트 시스템은 개별 에이전트가 단독으로 수행할 때보다 **시너지 효과(synergistic outcome)** 를 통해 더 우수한 성능을 발휘할 수 있다.

## 다중 에이전트 협업 패턴 개요 (Multi-Agent Collaboration Pattern Overview)

### 1. 개념
**다중 에이전트 협업 패턴(Multi-Agent Collaboration Pattern)** 은 여러 개의 **에이전트(agent)** 가 공통의 목표를 달성하기 위해 협력하는 시스템 설계 방식이다.  
각 에이전트는 명확히 정의된 역할, 개별 목표, 그리고 서로 다른 도구나 지식 기반에 접근할 수 있는 권한을 가진다.  
이 패턴의 핵심은 **에이전트 간의 상호작용과 시너지**에 있다.

---

### 2. 협업 형태
다중 에이전트 협업은 다양한 형태로 구현될 수 있다.

#### 🔹 Sequential Handoffs (순차적 인계)
- 한 에이전트가 작업을 완료하면 그 결과를 다음 에이전트에게 전달하는 파이프라인 구조.  
- **Planning 패턴**과 유사하지만, 각 단계가 **서로 다른 에이전트**에 의해 수행된다는 점이 다르다.

#### 🔹 Parallel Processing (병렬 처리)
- 여러 에이전트가 문제의 서로 다른 부분을 동시에 처리하고, 결과를 통합한다.  
- 처리 속도 향상과 병렬적 사고를 통한 문제 해결에 유리하다.

#### 🔹 Debate and Consensus (토론과 합의)
- 다양한 관점을 가진 에이전트들이 토론을 통해 선택지를 평가하고,  
  **합의(consensus)** 혹은 **더 정교한 결론**에 도달하는 구조.

#### 🔹 Hierarchical Structures (계층적 구조)
- **관리자(manager) 에이전트**가 하위 **작업자(worker) 에이전트**에게 작업을 동적으로 분배.  
- 각 에이전트는 자신이 다루는 도구나 플러그인 범위 내에서 문제를 처리한다.

#### 🔹 Expert Teams (전문가 팀)
- 각기 다른 전문성을 가진 에이전트들(예: 연구자, 작가, 편집자)이 협력하여 복잡한 결과물을 생성한다.

#### 🔹 Critic-Reviewer (비평가–검토자 구조)
- 일부 에이전트가 계획, 초안, 답변 등 **초기 산출물**을 생성하고,  
  다른 에이전트 그룹이 이를 **정책, 보안, 컴플라이언스, 품질, 정확성** 등의 기준으로 비판적으로 검토한다.  
- 이후 원 작성자나 최종 에이전트가 피드백을 반영해 결과를 수정한다.  
- 특히 **코드 생성, 연구 보고서 작성, 논리 검증, 윤리적 정합성 보장** 등에 효과적이다.  
- 이 접근법은 **강건성, 품질 향상, 환각(hallucination)이나 오류 감소** 등의 장점을 가진다.

---

### 3. 시스템 구조 요소
다중 에이전트 시스템은 다음 세 가지 핵심 요소로 구성된다.
1. **에이전트의 역할 및 책임 정의**
2. **정보 교환을 위한 통신 채널 구축**
3. **협업을 지시하는 작업 흐름(task flow) 또는 상호작용 프로토콜 설계**

## 다중 에이전트 협업: 상호 관계와 통신 구조 탐구  
(Multi-Agent Collaboration: Exploring Interrelationships and Communication Structures)

---

### 2. 주요 상호 관계 및 통신 모델




#### 🔹 ① 단일 에이전트 (Single Agent)
- 하나의 에이전트가 외부와의 통신 없이 자율적으로 작동한다.  
- 구현이 간단하고 관리가 용이하지만, 개별 에이전트의 자원과 능력에 한정된다.  
- **서로 독립적인 하위 문제로 분해 가능한 작업**에 적합하다.

---

#### 🔹 ② 네트워크 모델 (Network)
- 여러 에이전트가 **분산된 방식으로(peer-to-peer)** 직접 통신하며 협력한다.  
- 정보, 자원, 작업을 상호 교환할 수 있다.  
- 특정 에이전트의 실패가 전체 시스템에 치명적 영향을 주지 않아 **탄력성(resilience)** 이 높다.  
- 그러나 통신량이 증가하고, 일관된 의사결정 유지가 어려울 수 있다.

---

#### 🔹 ③ 감독자 모델 (Supervisor)
- **중앙 관리자(Supervisor)** 에이전트가 하위 에이전트들의 작업을 조율하고 관리한다.  
- 명확한 권한 구조를 제공하여 관리가 용이하나,  
  **단일 실패 지점(single point of failure)** 이 발생할 수 있고,  
  감독자가 과부하될 경우 병목 현상이 생길 수 있다.

---

#### 🔹 ④ 도구형 감독자 (Supervisor as a Tool)
- 감독자가 직접 지시하기보다는 **도구나 리소스 제공자** 역할을 수행한다.  
- 예: 데이터, 분석 도구, 계산 자원을 제공하여 다른 에이전트들이 더 효율적으로 작업할 수 있도록 지원.  
- 상향식 자율성을 보존하면서 **지능적 지원 기능**을 결합한 유연한 형태이다.

---

#### 🔹 ⑤ 계층형 모델 (Hierarchical)
- 다단계의 감독 구조로, 상위 감독자가 하위 감독자를 관리하고,  
  그 아래에서 여러 작업 에이전트들이 실제 업무를 수행한다.  
- **복잡한 문제를 계층적으로 분해**하여 처리할 수 있으며,  
  **확장성과 분산 의사결정 구조**를 모두 갖춘 체계적인 모델이다.

---

#### 🔹 ⑥ 맞춤형 모델 (Custom)
- 특정 문제나 응용 분야에 맞게 **맞춤 설계된(interrelationship & communication)** 구조.  
- 기존 모델의 하이브리드 형태 또는 완전히 새로운 통신 구조를 설계할 수 있다.  
- 예:  
  - 실시간 환경에서의 적응형 협업  
  - 도메인 지식 기반의 의사결정  
  - 성능 최적화 및 예외 상황 처리용 특화 모델  
- 구현에는 **심층적인 MAS 설계 원리 이해와 정교한 통신 프로토콜 설계 능력**이 요구된다.

### 3. 요약 및 시사점

  - 시스템의 강건성 요구  
  - 허용 가능한 통신 오버헤드

```python
from __future__ import annotations

import os
from typing import TypedDict, List, Literal, Annotated

from dotenv import load_dotenv
from pydantic import BaseModel, Field

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, ToolMessage, BaseMessage
from langchain.tools import tool

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode
```

```python
load_dotenv()  # Load environment variables from .env file

BASE_URL = os.getenv("OPENAI_BASE_URL")
API_KEY = os.getenv("OPENAI_API_KEY")
MODEL_NAME = os.getenv("MODEL_NAME")

llm = ChatOpenAI(
    model_name=MODEL_NAME,
    base_url=BASE_URL,
    api_key=API_KEY,
    temperature=0.3,
)
```

```python
llm.invoke([HumanMessage(content="Hello World")])  # Test the LLM connection
```

```python
from tavily import TavilyClient

_tavily = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))

class SearchInput(BaseModel):
    query: str = Field(..., description="웹 검색 질의")
    max_results: int = Field(5, ge=1, le=10, description="최대 검색 결과 수")


@tool(args_schema=SearchInput)
def web_search(query: str, max_results: int = 5) -> List[dict]:
    """
    간단한 웹 검색. Tavily 키가 없으면 모의 결과를 반환합니다.
    반환: [{url, title, content}, ...]
    """
    res = _tavily.search(query=query, max_results=max_results)
    return res.get("results", [])

tool_node = ToolNode([web_search])
```

```python
def researcher_llm(state: State) -> dict:
    """연구자 역할. 필요 시 web_search tool-call을 스스로 결정."""
    system = (
        "당신은 시니어 리서치 애널리스트입니다.\n"
        "- 주제와 대상 독자에 맞춰 2024–2025년 AI 트렌드 3~5개를 제시하세요.\n"
        "- 필요할 때만 web_search 도구를 1~2회 이내로 호출하여 근거 URL을 확보하세요.\n"
        "- 출력은 한국어 불릿 요약(5~8개) + 3~5문장 핵심 요약으로 간결하게 작성하세요."
    )
    user = (
        f"주제: {state['topic']}\n"
        f"대상 독자: {state['audience']}\n"
        "길게 작성하지 말고, 근거가 있다면 마지막에 URL을 함께 남겨주세요."
    )
    llm_with_tools = llm.bind_tools([web_search])
    ai = llm_with_tools.invoke([
        HumanMessage(content=system),
        HumanMessage(content=user)
    ])
    return {"messages": [ai], "research_summary": ai.content}

def should_call_tool(state: State) -> Literal["call_tools", "to_writer"]:
    """마지막 AI 메시지에 tool_calls가 있으면 ToolNode로, 없으면 작가 단계로 분기."""
    last = next((m for m in reversed(state["messages"]) if isinstance(m, AIMessage)), None)
    loops = int(state.get("tool_loops", 0))
    if last and getattr(last, "tool_calls", None) and loops < 2:
        return "call_tools"
    return "to_writer"

def inc_tool_loops(state: State) -> dict:
    """tool_loops 카운터 1 증가."""
    loops = int(state.get("tool_loops", 0)) + 1
    return {"tool_loops": loops}

def writer_llm(state: State) -> dict:
    """작가: 지금까지의 대화(연구 요약/검색 결과)를 바탕으로 500~700자 한국어 블로그 글 작성."""
    prompt = (
        "당신은 테크니컬 라이터입니다.\n"
        "- 지금까지의 연구 요약과 검색 결과를 바탕으로 한국어 블로그 글(500~700자)을 작성하세요.\n"
        "- 소제목 2~3개, 불릿 소량을 포함하고, 과장 없이 명확하고 간결하게 작성하세요."
        f"연구 요약:\n{state['research_summary']}\n"
    )
    ai = llm.invoke([HumanMessage(content=prompt)])
    print(f"writer output : {ai.content}")
    return {"messages": [ai], "blog_post": ai.content}
```

```python
graph = StateGraph(State)
graph.add_node("researcher_llm", researcher_llm)
graph.add_node("tools", tool_node)
graph.add_node("inc_tool_loops", inc_tool_loops)
graph.add_node("writer_llm", writer_llm)

graph.add_edge(START, "researcher_llm")
graph.add_conditional_edges(
    "researcher_llm",
    should_call_tool,
    {
        "call_tools": "tools",
        "to_writer": "writer_llm",
    }
)
graph.add_edge("tools", "inc_tool_loops")
graph.add_edge("inc_tool_loops", "researcher_llm")
graph.add_edge("writer_llm", END)

app = graph.compile()
```

## Hierarchical Agent Structure

```python
from __future__ import annotations

import os
import re
from typing import TypedDict, List, Literal, Annotated, Optional

from dotenv import load_dotenv
from pydantic import BaseModel, Field

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, BaseMessage, SystemMessage
from langchain.tools import tool

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
```

```python
load_dotenv()

BASE_URL = os.getenv("OPENAI_BASE_URL")
API_KEY = os.getenv("OPENAI_API_KEY")
MODEL_NAME = os.getenv("MODEL_NAME")
```

```python
llm = ChatOpenAI(
    model=MODEL_NAME,
    base_url=BASE_URL,
    api_key=API_KEY,
    temperature=0.3,
)
```

```python
class RouteArgs(BaseModel):
    rationale: Optional[str] = Field(None, description="이 라우팅 결정을 내린 이유")

@tool("route_to_greeter", args_schema=RouteArgs)
def route_to_greeter(rationale: Optional[str] = None) -> str:
    """Greeter 하위 에이전트에게 위임하라는 라우팅 신호."""
    return "ROUTE_GREETER"

@tool("route_to_task", args_schema=RouteArgs)
def route_to_task(rationale: Optional[str] = None) -> str:
    """TaskExecutor 하위 에이전트에게 위임하라는 라우팅 신호."""
    return "ROUTE_TASK"

ROUTING_TOOLS = [route_to_greeter, route_to_task]
```

```python
class State(TypedDict):
    user_input: str
    messages: Annotated[List[BaseMessage], add_messages]
    route: Optional[Literal["greeter", "task"]]
```

```python
def coordinator_llm(state: State) -> dict:
    system = (
        "당신은 상위 코디네이터(Coordinator)입니다.\n"
        "- 사용자의 요청을 읽고, 아래 두 도구 중 정확히 하나만 호출하여 위임하세요.\n"
        "  1) route_to_greeter: 인사/환영/소개/가벼운 응대가 필요한 경우\n"
        "  2) route_to_task   : 구체적 작업 실행/처리(비-LLM 태스크)가 필요한 경우\n"
        "- 도구 호출 외의 텍스트 출력은 하지 마세요."
    )
    user = f"[사용자 요청]\n{state['user_input']}"
    llm_with_tools = llm.bind_tools(ROUTING_TOOLS)
    message = llm_with_tools.invoke([
        HumanMessage(content=system),
        HumanMessage(content=user)
    ])
    return {"messages": [message]}
```

```python
def should_route(state: State) -> Literal["to_greeter", "to_task"]:
    last_ai_message = None
    for m in reversed(state["messages"]):
        if isinstance(m, AIMessage):
            last_ai_message = m
            break
    if last_ai_message and getattr(last_ai_message, "tool_calls", None):
        calls = [c["function"]["name"] for c in last_ai_message.tool_calls if "function" in c]
        if "route_to_greeter" in calls:
            return "to_greeter"
        elif "route_to_task" in calls:
            return "to_task"

    return "to_greeter"  # 기본값
```

```python
def greeter_llm(state: State) -> dict:
    system = (
        "당신은 친근한 인사 담당 에이전트(Greeter)입니다.\n"
        "- 사용자에게 공손하고 따뜻하게 인사하세요.\n"
        "- 2~4문장, 과장 금지, 한국어."
    )
    user = f"[사용자 요청]\n{state['user_input']}"
    message = llm.invoke([
        SystemMessage(content=system),
        HumanMessage(content=user)
    ])
    return {"messages": [message]}
```

```python
def task_executor_node(state: State) -> dict:
    event_text = "Task finished successfully."
    ai_like = AIMessage(content=event_text, name="TaskExecutor")
    return {"messages": [ai_like]}
```

```python
graph = StateGraph(State)
graph.add_node("coordinator_llm", coordinator_llm)
graph.add_node("greeter_llm", greeter_llm)
graph.add_node("task_executor", task_executor_node)

graph.add_edge(START, "coordinator_llm")
graph.add_conditional_edges(
    "coordinator_llm",
    should_route,
    {
        "to_greeter": "greeter_llm",
        "to_task": "task_executor",
    }
)
graph.add_edge("greeter_llm", END)
graph.add_edge("task_executor", END)

app = graph.compile()
```

## LoopAgent 패턴 : 계층형 반복 오케스트레이션 패턴

```python
from __future__ import annotations

import os, re
from typing import TypedDict, List, Literal, Annotated, Optional

from dotenv import load_dotenv
from pydantic import BaseModel, Field

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, ToolMessage, BaseMessage
from langchain.tools import tool

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode
```

```python
load_dotenv()

BASE_URL = os.getenv("OPENAI_BASE_URL")
API_KEY = os.getenv("OPENAI_API_KEY")
MODEL_NAME = os.getenv("MODEL_NAME")
```

```python
llm = ChatOpenAI(
    model=MODEL_NAME,
    base_url=BASE_URL,
    api_key=API_KEY,
    temperature=0.3,
)
```

```python
class State(TypedDict):
    goal: str                                 # 전체 목표(사용자 의도)
    messages: Annotated[List[BaseMessage], add_messages]  # 대화/툴콜 로그
    status: Literal["pending", "completed"]   # 진행 상태
    loop_count: int                            # 현재 반복 횟수(0부터)
    max_iters: int                             # 최대 반복 상한
```

```python
class SetStatusArgs(BaseModel):
    status: Literal["pending", "completed"] = Field(..., description="설정할 상태")

@tool("set_status", args_schema=SetStatusArgs)
def set_status(status: Literal["pending", "completed"]) -> str:
    """상태를 변경하라는 신호를 반환 (예: 'STATUS=completed'). 실제 반영은 별도 노드에서 수행."""
    return f"STATUS={status}"

tool_node = ToolNode([set_status])
```

```python
tool_node
```

```python
def processing_step(state: State) -> dict:
    system = (
        "당신은 'ProcessingStep'입니다.\n"
        "- 목표(goal)를 향해 한 번에 한 단계(step)만 수행하세요.\n"
        "- 단계 결과를 한국어 3~6문장으로 설명.\n"
        "- 이번 단계가 '최종 단계'라고 확신되면 반드시 set_status(status='completed') 도구 호출.\n"
        "- 확신이 없으면 도구 호출하지 말고 다음 반복을 위한 후속 작업 제안.\n"
        "- 과장 금지, 명확·구체적으로."
    )
    user = (
        f"[goal]\n{state['goal']}\n\n"
        f"[현재 상태] status={state['status']}, loop_count={state['loop_count']}/{state['max_iters']}\n"
        "주의: 정말 최종 단계라고 판단될 때만 set_status(status='completed')를 호출하세요."
    )
    llm_with_tools = llm.bind_tools([set_status])
    ai_message = llm_with_tools.invoke([
        HumanMessage(content=system),
        HumanMessage(content=user)
    ])
    return {"messages": [ai_message]}
```

```python
def apply_status(state: State) -> dict:
    new_status = None
    for m in reversed(state["messages"]):
        if isinstance(m, ToolMessage):
            text = (m.content or "").strip()
            hit = re.search(r"STATUS\s*=\s*(pending|completed)", text, re.IGNORECASE)
            if hit:
                new_status = hit.group(1).lower()
                break
    if new_status and new_status != state["status"]:
        note = AIMessage(content=f"[apply_status] '{state['status']}' → '{new_status}' 로 반영")
        return {"messages": [note], "status": new_status}
    return {"messages": [AIMessage(content="[apply_status] 변경 없음")]}
```

```python
def condition_checker(state: State) -> dict:
    status = state["status"]
    if status == "completed":
        msg = "조건 충족: status=completed → 루프 종료"
    else:
        remain = state["max_iters"] - state["loop_count"]
        msg = f"조건 미충족: status={status} → 루프 지속 (남은 반복 가능 횟수 {remain})"
    return {"message": [AIMessage(content=f"[ConditionChecker] {msg}", name="ConditionChecker")]}
```

```python
def should_continue(state: State) -> Literal["loop", "stop"]:
    if state["status"] == "completed":
        return "stop"
    if state["loop_count"] + 1 >= state["max_iters"]:
        return "stop"
    return "loop"

def inc_loop(state: State) -> dict:
    return {"loop_count": state["loop_count"] + 1,
            "messages": [AIMessage(content=f"[loop] loop_count → {state['loop_count'] + 1}")]}
```

```python
def should_call_tools(state: State) -> Literal["call_tools", "skip_tools"]:
    last_ai_message = None
    for m in reversed(state["messages"]):
        if isinstance(m, AIMessage):
            last_ai_message = m
            break
    if last_ai_message and getattr(last_ai_message, "tool_calls", None):
        return "call_tools"
    return "skip_tools"
```

```python
graph = StateGraph(State)
graph.add_node("processing_step", processing_step)
graph.add_node("tools", tool_node)
graph.add_node("apply_status", apply_status)
graph.add_node("condition_checker", condition_checker)
graph.add_node("inc_loop", inc_loop)

graph.add_edge(START, "processing_step")
graph.add_conditional_edges(
    "processing_step",
    should_call_tools,
    {
        "call_tools": "tools",
        "skip_tools": "apply_status",
    }
)
graph.add_edge("tools", "apply_status")
graph.add_edge("apply_status", "condition_checker")
graph.add_conditional_edges(
    "condition_checker",
    should_continue,
    {
        "loop": "inc_loop",
        "stop": END,
    }
)
graph.add_edge("inc_loop", "processing_step")

app = graph.compile()
```

```python
from IPython.display import Image, display

display(Image(app.get_graph().draw_mermaid_png()))
```

```python
init: State = {
    "goal": "데이터 전처리 파이프라인을 단계적으로 구축하라. 마지막 단계에 도달하면 completed로 설정하라.",
    "messages": [],
    "status": "pending",
    "loop_count": 0,
    "max_iters": 5,
}

out = app.invoke(init)

# 2) 최종 상태 요약
print("\n=== 최종 상태 ===")
print({k: out[k] for k in ("status", "loop_count")})
```

## 슈퍼바이저(오케스트레이터)가 있는 팬아웃/팬인(fan-out/fan-in)

```python
from __future__ import annotations

import os, json
from typing import TypedDict, List, Literal, Annotated

from dotenv import load_dotenv
from pydantic import BaseModel, Field

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, ToolMessage, BaseMessage
from langchain.tools import tool

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode
```

```python
load_dotenv()

BASE_URL = os.getenv("OPENAI_BASE_URL")
API_KEY = os.getenv("OPENAI_API_KEY")
MODEL_NAME = os.getenv("MODEL_NAME")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model=MODEL_NAME,
    base_url=BASE_URL,
    api_key=API_KEY,
    temperature=0.2,
    timeout=90,
    # llama.cpp/llama-server는 바디 필드 n_predict를 인식
    extra_body={"n_predict": 256},
)
```

```python
class Plan(TypedDict, total=False):
    need_weather: bool
    need_news: bool
    rationale: str

class State(TypedDict):
    user_input: str
    location: str
    topic: str
    messages: Annotated[List[BaseMessage], add_messages]

    plan: Plan
    # 결과 저장
    weather_data: str
    news_data: str
    # 완료 플래그 (팬인용)
    weather_done: bool
    news_done: bool
    # 분기별 툴 호출 판정(경합 방지용)
    pending_weather_tool: bool
    pending_news_tool: bool
    # 최종 보고서
    report: str
```

```python
class WeatherArgs(BaseModel):
    location: str = Field(..., description="날씨 조회 위치(예: Seoul")

class NewsArgs(BaseModel):
    topic: str = Field(..., description="뉴스 토픽(예: AI, 스포츠)")

@tool("weather_api", args_schema=WeatherArgs)
def weather_api(location: str) -> str:
    """주어진 위치의 날씨 리포트를 반환(모의)."""
    return json.dumps(
        {"tool": "weather_api", "location": location, "report": f"{location}: 맑음, 25°C, 미세먼지 보통"},
        ensure_ascii=False,
    )

@tool("news_api", args_schema=NewsArgs)
def news_api(topic: str) -> str:
    """주어진 토픽의 탑 뉴스(모의)를 반환."""
    return json.dumps(
        {"tool": "news_api", "topic": topic, "headline": f"'{topic}' 최신 이슈 브리핑", "url": f"https://example.com/news?q={topic}"},
        ensure_ascii=False,
    )

tool_node_weather = ToolNode([weather_api])
tool_node_news = ToolNode([news_api])
```

```python
def _extract_tool_name_args(call):
    if call is None:
        return None, None
    if isinstance(call, dict):
        fn = call.get("function")
        if isinstance(fn, dict):  # 표준
            return fn.get("name"), fn.get("arguments")
        # 평평한 형태
        name = call.get("name") or call.get("tool_name") or call.get("function_name")
        args = call.get("arguments") or call.get("args") or call.get("kwargs")
        return name, args
    # 객체형 방어
    fn = getattr(call, "function", None)
    if fn is not None:
        return getattr(fn, "name", None), getattr(fn, "arguments", None)
    return getattr(call, "name", None), getattr(call, "arguments", None)
```

```python
def weather_llm(state: State) -> dict:
    plan = state.get("plna", {})
    if not plan.get("need_weather", False):
        print("[weather_llm] 스킵: 계획에 없음")
        return {"messages": [AIMessage(content="[weather_llm] skipped", name="WeatherAgent")], "weather_done": True}

    system = (
        "너는 '날씨 수집 에이전트'다.\n"
        "- 반드시 weather_api 도구 **한 번만** 호출해 주어진 location의 날씨만 가져와라.\n"
        "- 텍스트 설명은 출력하지 마라(툴콜만)."
    )

    user = f"[location]\n{state['location']}"
    llm_with_tool = llm.bind_tools([weather_api])
    ai = llm_with_tool.invoke(
        [
            HumanMessage(content=system),
            HumanMessage(content=user)
        ]
    )

    print("\n[weather_llm] 호출됨")
    pending = False

    tc = list(getattr(ai, "tool_calls", []) or [])
    if not tc:
        fc = (getattr(ai, "additional_kwargs", {}) or {}).get("function_call")
        if fc:
            tc = [fc]

    if tc:
        for i, call in enumerate(tc):
            name, args = _extract_tool_name_args(call)
            print(f"  - tool_call[{i}] name={name}, args={args}")  # ✅ f-string 적용
            if name == "weather_api":
                pending = True
    else:
        print("  - tool_calls 없음 (예상 외)")

    return {"messages": [ai], "pending_weather_tool": pending}

def should_call_weather_tools(state: State) -> Literal["call", "skip"]:
    return "call" if state.get("pending_weather_tool") else "skip"

def inspect_weather_tool(state: State) -> dict:
    print("[inspect_weather_tool] 호출됨")
    for m in reversed(state["messages"]):
        if isinstance(m, ToolMessage):
            print(f"  - ToolMessage(tool_call_id={getattr(m,'tool_call_id',None)}): {m.content}")
            break
    return {}

def apply_weather(state: State) -> dict:
    print("[apply_weather] 호출됨")
    for m in reversed(state["messages"]):
        if isinstance(m, ToolMessage) and (m.content or "").strip():
            try:
                obj = json.loads(m.content)
                if obj.get("tool") == "weather_api":
                    text = f"[날씨] {obj['report']}"
                    print("  - weather_data 반영:", text)
                    return {
                        "messages": [AIMessage(content=text, name="WeatherAgent")],
                        "weather_data": text,
                        "weather_done": True,
                        "pending_weather_tool": False,
                    }
            except Exception as e:
                print("  - JSON 파싱 실패:", e)
                break
    print("  - weather_api 결과 없음 → 완료표시만")
    return {"weather_done": True, "pending_weather_tool": False}
```

```python
def news_llm(state: State) -> dict:
    plan = state.get("plan", {})
    if not plan.get("need_news", False):
        print("[news_llm] 스킵: 계획에 없음")
        return {"messages": [AIMessage(content="[news_llm] skipped", name="NewsAgent")], "news_done": True}

    system = (
        "너는 '뉴스 수집 에이전트'다.\n"
        "- 반드시 news_api 도구 **한 번만** 호출해 주어진 topic의 탑 뉴스만 가져와라.\n"
        "- 텍스트 설명은 출력하지 마라(툴콜만)."
    )
    user = f"[topic]\n{state['topic']}"
    llm_with_tool = llm.bind_tools([news_api])
    ai = llm_with_tool.invoke([HumanMessage(content=system), HumanMessage(content=user)])

    print("\n[news_llm] 호출됨")
    pending = False

    tc = list(getattr(ai, "tool_calls", []) or [])
    if not tc:
        fc = (getattr(ai, "additional_kwargs", {}) or {}).get("function_call")
        if fc:
            tc = [fc]

    if tc:
        for i, call in enumerate(tc):
            name, args = _extract_tool_name_args(call)
            print(f"  - tool_call[{i}] name={name}, args={args}")  # ✅ f-string 적용
            if name == "weather_api":
                pending = True
    else:
        print("  - tool_calls 없음 (예상 외)")

    return {"messages": [ai], "pending_weather_tool": pending}



def should_call_news_tools(state: State) -> Literal["call", "skip"]:
    return "call" if state.get("pending_news_tool") else "skip"

def inspect_news_tool(state: State) -> dict:
    print("[inspect_news_tool] 호출됨")
    for m in reversed(state["messages"]):
        if isinstance(m, ToolMessage):
            print(f"  - ToolMessage(tool_call_id={getattr(m,'tool_call_id',None)}): {m.content}")
            break
    return {}

def apply_news(state: State) -> dict:
    print("[apply_news] 호출됨")
    for m in reversed(state["messages"]):
        if isinstance(m, ToolMessage) and (m.content or "").strip():
            try:
                obj = json.loads(m.content)
                if obj.get("tool") == "news_api":
                    text = f"[뉴스] {obj['headline']} — {obj['url']}"
                    print("  - news_data 반영:", text)
                    return {
                        "messages": [AIMessage(content=text, name="NewsAgent")],
                        "news_data": text,
                        "news_done": True,
                        "pending_news_tool": False,
                    }
            except Exception as e:
                print("  - JSON 파싱 실패:", e)
                break
    print("  - news_api 결과 없음 → 완료표시만")
    return {"news_done": True, "pending_news_tool": False}
```

```python
def barrier_join(state: State) -> dict:
    w_done = bool(state.get("weather_done"))
    n_done = bool(state.get("news_done"))
    print(f"[barrier_join] weather_done={w_done}, news_done={n_done}")
    return {}

def should_finalize(state: State) -> Literal["finalize", "wait"]:
    return "finalize" if (bool(state.get("weather_done")) and bool(state.get("news_done"))) else "wait"

def finalize_report(state: State) -> dict:
    print("[finalize_report] 호출됨 — 팬인 완료, 최종 리포트 생성")
    weather = state.get("weather_data", "").strip()
    news    = state.get("news_data", "").strip()

    parts = []
    if weather: parts.append(weather)
    if news:    parts.append(news)
    if not parts:
        parts.append("수집된 데이터가 없습니다.")

    report = " | ".join(parts)
    msg = AIMessage(content=f"[Final]\n{report}")
    return {"messages": [msg], "report": report}

def noop(state: State) -> dict:
    # 팬인 대기 분기: 다른 분기가 barrier_join에 도달할 때까지 이 분기는 종료
    print("[noop] 대기 분기 — 다른 분기의 완료를 기다리는 중")
    return {}
```

```python
graph = StateGraph(State)

# Supervisor
graph.add_node("supervisor", supervisor_node)

# Weather branch
graph.add_node("weather_llm", weather_llm)
graph.add_node("tools_weather", tool_node_weather)
graph.add_node("inspect_weather_tool", inspect_weather_tool)
graph.add_node("apply_weather", apply_weather)

# News branch
graph.add_node("news_llm", news_llm)
graph.add_node("tools_news", tool_node_news)
graph.add_node("inspect_news_tool", inspect_news_tool)
graph.add_node("apply_news", apply_news)

# Barrier & Finalizer
graph.add_node("barrier_join", barrier_join)
graph.add_node("finalize_report", finalize_report)
graph.add_node("noop", noop)

# START → Supervisor
graph.add_edge(START, "supervisor")

# Fan-out: Supervisor → (Weather, News) 병렬 시작
graph.add_edge("supervisor", "weather_llm")
graph.add_edge("supervisor", "news_llm")

# Weather flow
graph.add_conditional_edges("weather_llm", should_call_weather_tools, {
    "call": "tools_weather",
    "skip": "apply_weather",
})
graph.add_edge("tools_weather", "inspect_weather_tool")
graph.add_edge("inspect_weather_tool", "apply_weather")
graph.add_edge("apply_weather", "barrier_join")

# News flow
graph.add_conditional_edges("news_llm", should_call_news_tools, {
    "call": "tools_news",
    "skip": "apply_news",
})
graph.add_edge("tools_news", "inspect_news_tool")
graph.add_edge("inspect_news_tool", "apply_news")
graph.add_edge("apply_news", "barrier_join")

# Fan-in: 두 갈래가 barrier_join으로 모임 → 조건에 따라 finalize 또는 대기
graph.add_conditional_edges("barrier_join", should_finalize, {
    "finalize": "finalize_report",
    "wait": "noop",
})
graph.add_edge("finalize_report", END)
# 'noop'은 다른 분기가 barrier_join에 도달할 때까지 대기(출구 엣지 없음)

app = graph.compile()
```

## 상위 Artist(오케스트레이터) 가 하위 ImageGen 서브에이전트(별도 그래프) 를 툴처럼 호출

```python
from __future__ import annotations
import os, json, base64
from typing import TypedDict, List, Literal, Annotated

from dotenv import load_dotenv
from pydantic import BaseModel, Field

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage, ToolMessage, BaseMessage
from langchain.tools import tool

from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode
```

```python
load_dotenv()

BASE_URL = os.getenv("OPENAI_BASE_URL")
API_KEY = os.getenv("OPENAI_API_KEY")
MODEL_NAME = os.getenv("MODEL_NAME")
```

```python
llm = ChatOpenAI(
    model=MODEL_NAME,
    base_url=BASE_URL,
    api_key=API_KEY,
    temperature=0.2,
    extra_body={"n_predict": 256},
)
```

```python
def _extract_tool_name_args(call):
    """tool_calls의 원소에서 (name, arguments:str|dict) 안전 추출."""
    if call is None:
        return None, None
    if isinstance(call, dict):
        fn = call.get("function")
        if isinstance(fn, dict):  # OpenAI 표준형
            return fn.get("name"), fn.get("arguments")
        # 평평한 형태
        name = call.get("name") or call.get("tool_name") or call.get("function_name")
        args = call.get("arguments") or call.get("args") or call.get("kwargs")
        return name, args
    # 객체형(래퍼)
    fn = getattr(call, "function", None)
    if fn is not None:
        return getattr(fn, "name", None), getattr(fn, "arguments", None)
    return getattr(call, "name", None), getattr(call, "arguments", None)

def _as_json(obj_or_str, default=None):
    if obj_or_str is None:
        return default
    if isinstance(obj_or_str, (dict, list)):
        return obj_or_str
    try:
        return json.loads(obj_or_str)
    except Exception:
        return default
```

```python
class ImageState(TypedDict):
    prompt: str
    messages: Annotated[List[BaseMessage], add_messages]
    iamge_b64: str
    image_mime: str

class GenArgs(BaseModel):
    prompt: str = Field(..., description="생성할 이미지 프롬프트")

@tool("generate_image", args_schema=GenArgs)
def generate_image(prompt: str) -> str:
    """
    (데모) 이미지 생성 툴. 실제에선 이미지 API 연동.
    여기서는 mock 바이트를 base64로 반환.
    """
    print(f"[TOOL/generate_image] 프롬프트 수신: {prompt}")
    mock_image_bytes = f"IMAGE_BYTES_FOR_PROMPT:{prompt}".encode("utf-8")
    payload = {"status": "success", "mime": "image/png", "b64": base64.b64encode(mock_image_bytes).decode("utf-8")}
    return json.dumps(payload, ensure_ascii=False)

_image_tool_node = ToolNode([generate_image])

def imagegen_llm(state: ImageState) -> dict:
    """이미지 생성 전문가: 반드시 generate_image 툴을 1회 호출."""
    system = (
        "너는 '이미지 생성 전문가'다.\n"
        "- 사용자가 준 전체 요청을 그대로 'prompt' 인자로 하여 generate_image 도구를 정확히 1회 호출하라.\n"
        "- 도구 호출 외의 텍스트를 출력하지 마라."
    )
    user = f"[prompt]\n{state['prompt']}"
    llm_with_tool = llm.bind_tools([generate_image], parallel_tool_calls=False)
    ai = llm_with_tool.invoke([
        HumanMessage(content=system),
        HumanMessage(content=user)
    ])

    print("\n[ImageGen/imagegen_llm] 호출됨")
    tc = list(getattr(ai, "tool_calls", []) or [])
    if not tc:
        fc = (getattr(ai, "additional_kwargs", {}) or {}).get("function_call")
        if fc:
            tc = [fc]

    if tc:
        for i, call in enumerate(tc):
            name, args = _extract_tool_name_args(call)
            print(f"  - tool_call[{i}] name={name}, args={args}")  # ✅ f-string 적용
    else:
        print("  - tool_calls 없음 (예상 외)")

    return {"messages": [ai]}

def should_call_image_tool(state: ImageState) -> Literal["call", "skip"]:
    last_ai = next((m for m in reversed(state["messages"]) if isinstance(m, AIMessage)), None)
    if isinstance(last_ai, AIMessage):
        tc = list(getattr(last_ai, "tool_calls", []) or [])
        if not tc:
            fc = (getattr(last_ai, "additional_kwargs", {}) or {}).get("function_call")
            if fc:
                tc = [fc]
        for call in tc:
            name, _ = _extract_tool_name_args(call)
            if name == "generate_image":
                return "call"
    return "skip"

def apply_image_result(state: ImageState) -> dict:
    """ToolMessage에서 base64/MIME 추출하여 상태 반영."""
    print("[ImageGen/apply_image_result] 호출됨")
    b64, mime = "", "image/png"
    for m in reversed(state["messages"]):
        if isinstance(m, ToolMessage) and (m.content or "").strip():
            obj = _as_json(m.content, {})
            if obj and obj.get("status") == "success" and obj.get("b64"):
                b64 = obj["b64"]
                mime = obj.get("mime", "image/png")
                break
    if b64:
        print("  - 이미지 base64 수신(길이):", len(b64))
        note = AIMessage(content=f"[ImageGen] 이미지 생성 완료 (MIME: {mime}, 길이: {len(b64)})", name="ImageGen")
        return {"messages": [note], "iamge_b64": b64, "image_mime": mime}
    print("  - 이미지 결과 없음")
    return {"messages": [AIMessage(content="[ImageGen] 결과 없음")]}

_img_graph = StateGraph(ImageState)
_img_graph.add_node("imagegen_llm", imagegen_llm)
_img_graph.add_node("tools_image", _image_tool_node)
_img_graph.add_node("apply_image_result", apply_image_result)

_img_graph.add_edge(START, "imagegen_llm")
_img_graph.add_conditional_edges("imagegen_llm", should_call_image_tool, {
    "call": "tools_image",
    "skip": "apply_image_result",
})
_img_graph.add_edge("tools_image", "apply_image_result")
_img_graph.add_edge("apply_image_result", END)

IMAGE_AGENT = _img_graph.compile()
```

```python
class ImageAgentToolArgs(BaseModel):
    prompt: str = Field(..., description="하위 에이전트에 전달할 프롬프트")

@tool("imagegen_agent_tool", args_schema=ImageAgentToolArgs)
def imagegen_agent_tool(prompt: str) -> str:
    """
    (Agent-as-Tool) 하위 ImageGen 에이전트를 호출하여 이미지를 생성하고,
    {status, mime, b64} JSON을 반환.
    """
    print("[TOOL/imagegen_agent_tool] 하위 에이전트 호출 시작")
    init: ImageState = {"prompt": prompt, "messages": [], "image_b64": "", "image_mime": "image/png"}
    out = IMAGE_AGENT.invoke(init)
    payload = {"status": "ok", "mime": out.get("image_mime", "image/png"), "b64": out.get("image_b64", "")}
    print(f"[TOOL/imagegen_agent_tool] 완료, b64_len={len(payload['b64'])}")
    return json.dumps(payload, ensure_ascii=False)

_top_tool_node = ToolNode([imagegen_agent_tool])
```

```python
class TopState(TypedDict):
    task: str
    messages: Annotated[List[BaseMessage], add_messages]
    image_b64: str
    image_mime: str
    prompt_used: str

def artist_llm(state: TopState) -> dict:
    system = (
        "너는 '크리에이티브 아티스트'다.\n"
        "- 먼저 매우 창의적이고 구체적인 이미지 프롬프트를 1개 발명하라.\n"
        "- 그 프롬프트를 인자로 하여 반드시 imagegen_agent_tool을 1회 호출하라.\n"
        "- 텍스트 설명은 최소화하고, 툴콜을 중심으로 수행하라."
    )
    user = f"[task]\n{state['task']}"
    llm_with_tool = llm.bind_tools([imagegen_agent_tool], parallel_tool_calls=False)
    ai = llm_with_tool.invoke([HumanMessage(content=system), HumanMessage(content=user)])

    print("\n[Artist/artist_llm] 호출됨")
    prompt_used = ""
    tc = list(getattr(ai, "tool_calls", []) or [])
    if not tc:
        fc = (getattr(ai, "additional_kwargs", {}) or {}).get("function_call")
        if fc:
            tc = [fc]
    if tc:
        for i, call in enumerate(tc):
            name, args = _extract_tool_name_args(call)
            print(f"  - tool_call[{i}] name={name}, args={args}")
            if name == "imagegen_agent_tool":
                args_obj = _as_json(args, {})
                prompt_used = (args_obj or {}).get("prompt", "")
    else:
        print("  - tool_calls 없음 (예상 외)")

    return {"messages": [ai], "prompt_used": prompt_used}

def should_call_top_tool(state: TopState) -> Literal["call", "skip"]:
    last_ai = next((m for m in reversed(state["messages"]) if isinstance(m, AIMessage)), None)
    if isinstance(last_ai, AIMessage):
        tc = list(getattr(last_ai, "tool_calls", []) or [])
        if not tc:
            fc = (getattr(last_ai, "additional_kwargs", {}) or {}).get("function_call")
            if fc:
                tc = [fc]
        for call in tc:
            name, _ = _extract_tool_name_args(call)
            if name == "imagegen_agent_tool":
                return "call"
    return "skip"

def apply_top_image(state: TopState) -> dict:
    """imagegen_agent_tool의 ToolMessage 결과를 TopState에 반영."""
    print("[Artist/apply_top_image] 호출됨")
    b64, mime = "", "image/png"
    for m in reversed(state["messages"]):
        if isinstance(m, ToolMessage) and (m.content or "").strip():
            obj = _as_json(m.content, {})
            if obj and obj.get("status") in ("ok", "success") and obj.get("b64"):
                b64 = obj["b64"]; mime = obj.get("mime", "image/png")
                break
    if b64:
        print("  - 최종 이미지 수신(b64_len):", len(b64))
        note = AIMessage(content=f"[Artist] 이미지 생성 완료 (prompt='{state.get('prompt_used','')[:60]}...')")
        return {"messages": [note], "image_b64": b64, "image_mime": mime}
    print("  - Tool 결과 없음")
    return {"messages": [AIMessage(content='[Artist] 결과 없음')]}

# Top 그래프 컴파일
_top = StateGraph(TopState)
_top.add_node("artist_llm", artist_llm)
_top.add_node("top_tools", _top_tool_node)
_top.add_node("apply_top_image", apply_top_image)
_top.add_edge(START, "artist_llm")
_top.add_conditional_edges("artist_llm", should_call_top_tool, {
    "call": "top_tools",
    "skip": "apply_top_image",
})
_top.add_edge("top_tools", "apply_top_image")
_top.add_edge("apply_top_image", END)

APP = _top.compile()
```
