# Chapter 11: Goal Setting and Monitoring

AI 에이전트가 진정으로 효과적이고 목적에 부합하기 위해서는 단순히 정보를 처리하거나 도구를 사용하는 능력만으로는 충분하지 않습니다.  
에이전트에게는 **명확한 방향성**이 필요하며, 또한 **스스로 성공하고 있는지 판단할 수 있는 체계**가 필요합니다.

**목표 설정 및 모니터링(Goal Setting and Monitoring)** 패턴이 그 역할을 한다.

이 패턴은 에이전트에게 다음을 제공합니다:

- **구체적이고 명시적인 목표(objectives)**  
- **목표 달성 여부를 추적할 수 있는 수단(progress tracking)**  
- **현재 상태를 평가하고 조정할 수 있는 기준(success criteria)**  

## Goal Setting and Monitoring 패턴 개요

여행 계획을 세우는 상황을 떠올려 보세요. 목적지에 갑자기 순간이동하듯 도착하는 것이 아닙니다.  
먼저 **어디로 갈지(목표 상태)**를 정하고, **현재 어디에 있는지(초기 상태)**를 파악하며,  
**가능한 옵션(교통수단, 경로, 예산 등)**을 고려한 다음 다음과 같은 **단계적 절차**를 세웁니다:

- 티켓 예약  
- 짐 꾸리기  
- 공항/역으로 이동  
- 탑승  
- 도착  
- 숙소 찾기  

이처럼 의존성과 제약 조건을 고려하며 단계별로 계획을 세우는 과정이  
바로 **에이전트 시스템에서 말하는 ‘계획(Planning)’**입니다.

---

AI 에이전트의 맥락에서 ‘계획’은 다음을 의미합니다:

- 에이전트가 **고수준 목표(high-level objective)**를 입력받으면  
- 스스로 혹은 반자동적으로 **중간 단계(step) 또는 하위 목표(sub-goal)**를 만들어 내며  
- 이를 **순차적** 또는 **더 복잡한 플로우**로 실행하는 과정

이 과정에서 에이전트는 다음과 같은 다른 패턴과도 연계될 수 있습니다:

- 도구 사용 패턴 (tool use)  
- 라우팅 패턴 (routing)  
- 멀티 에이전트 협력 (multi-agent collaboration)

계획을 생성하는 방식은 다양할 수 있습니다:

- 정교한 검색 알고리즘  
- 논리 기반 추론  
- 또는 최근에는 LLM이 학습 데이터와 태스크에 대한 이해에 기반해  
  **가능성 있고 효과적인 계획을 생성**하는 방식이 널리 활용됩니다.

---

강력한 계획 능력을 갖춘 에이전트는 다음을 수행할 수 있습니다:

- 단일 질문으로 해결되지 않는 **복잡한 문제** 해결  
- 요청이 다면적일 때 **여러 측면을 동시에 고려**  
- 상황 변화에 따라 **재계획(Replanning)**  
- **복잡한 워크플로우**를 구성하고 실행  
- 단순 반응형 시스템을 넘어  
  **명확한 목표를 향해 능동적으로 움직이는 시스템**으로 발전

즉, Goal Setting & Monitoring 패턴은  
에이전트가 목적지까지 가는 길을 스스로 설계하고, 추적하고, 조정하도록 만드는  
고급 에이전트 행동의 핵심 기반이라고 할 수 있습니다.

## Practical Applications & Use Cases

복잡한 현실 환경에서 **자율적이고 신뢰성 있게 동작하는 에이전트**를 만들기 위해 

### 1. 고객 지원 자동화 (Customer Support Automation)

- **목표:**  
  “고객의 청구(빌링) 관련 문의를 해결하라.”
- **에이전트 행동:**  
  - 대화 내용을 실시간 모니터링  
  - 데이터베이스 조회 (청구 내역, 결제 기록 등)  
  - 필요한 경우 청구 금액 조정 등의 도구 호출
- **성공 모니터링:**  
  - 청구 변경이 실제로 반영되었는지 확인  
  - 고객의 긍정적인 피드백/만족도 체크  
  - 문제가 해결되지 않으면 **에스컬레이션**

---

### 2. 개인화 학습 시스템 (Personalized Learning Systems)

- **목표:**  
  “학생의 대수학(algebra) 이해도를 향상시켜라.”
- **에이전트 행동:**  
  - 학생의 문제 풀이 진행 상황 모니터링  
  - 학생의 수준에 맞춰 학습 자료와 문제를 동적으로 조정  
  - 정확도, 완료 시간 등 **성과 지표(performance metrics)** 추적
- **성공 모니터링 및 적응:**  
  - 학생이 계속 어려워한다면 설명 방식·난이도·문제 유형을 변경  
  - 특정 개념에서 반복적으로 틀리면 보충 설명과 추가 연습 제공

---

### 3. 프로젝트 관리 어시스턴트 (Project Management Assistants)

- **목표:**  
  “프로젝트 마일스톤 X를 Y일까지 완료하라.”
- **에이전트 행동:**  
  - 개별 작업의 상태 및 진척률 모니터링  
  - 팀 커뮤니케이션(메일, 채팅, 이슈 트래커 등) 확인  
  - 인력·예산 등 자원 가용성 체크
- **성공 모니터링 및 개입:**  
  - 지연 가능성이 보이면 미리 경고  
  - 우선순위 조정, 인력 재배치 등 **교정 액션(corrective actions)** 제안

---

### 4. 자동화된 트레이딩 봇 (Automated Trading Bots)

- **목표:**  
  “위험 허용 범위 내에서 포트폴리오 수익을 최대화하라.”
- **에이전트 행동:**  
  - 실시간 시장 데이터, 뉴스, 지표 모니터링  
  - 현재 포트폴리오 가치와 개별 자산 비중 추적  
  - 사전에 정의된 전략과 목표에 맞춰 매수·매도 실행
- **성공 모니터링 및 전략 조정:**  
  - 손실·변동성이 설정된 위험 한도를 넘어서면 전략 수정  
  - 리스크 지표(예: VaR, 최대 낙폭 등)를 기반으로 포지션 축소/헤지 수행

---

### 5. 로보틱스 & 자율주행 차량 (Robotics and Autonomous Vehicles)

- **목표:**  
  “승객을 A 지점에서 B 지점까지 안전하게 이동시켜라.”
- **에이전트 행동:**  
  - 주변 환경(다른 차량, 보행자, 신호등 등)을 지속적으로 인지  
  - 차량 자체 상태(속도, 연료/배터리 등) 모니터링  
  - 계획된 경로 상에서의 진행 상황 추적
- **성공 모니터링 및 적응:**  
  - 위험 상황 또는 도로 상황 변화에 따라 속도·경로·운전 패턴 조정  
  - 안전과 효율(시간, 연료)을 모두 고려한 의사결정 수행

---

### 6. 콘텐츠 모더레이션 (Content Moderation)

- **목표:**  
  “플랫폼 X에서 유해 콘텐츠를 식별하고 제거하라.”
- **에이전트 행동:**  
  - 새로 업로드되는 콘텐츠(텍스트, 이미지, 영상 등) 모니터링  
  - 분류 모델을 활용해 규정 위반 여부 판별  
- **성공 모니터링 및 개선:**  
  - **오탐(false positive)**, **미탐(false negative)** 비율과 같은 지표 추적  
  - 기준이 너무 엄격/느슨하면 필터링 규칙이나 모델 파라미터 조정  
  - 애매한 사례는 인간 검토자에게 에스컬레이션

---

Goal Setting and Monitoring 패턴은 에이전트가 다음을 할 수 있게 해 줍니다:

- **신뢰도 있게 동작**하고  
- **구체적인 결과를 달성**하며  
- **동적인 환경 변화에 적응**하도록 돕는  
지능적인 **자기 관리(self-management)**의 필수 프레임워크입니다.


```python
import os
from dotenv import load_dotenv

load_dotenv()  # Load environment variables from .env file

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
GEMINI_MODEL = "gemini-2.5-flash"
```

```python
from __future__ import annotations

from typing import TypedDict, List, Optional

from langgraph.graph import StateGraph, START, END
from langgraph.checkpoint.memory import InMemorySaver

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage, BaseMessage
```

```python
llm = ChatGoogleGenerativeAI(
    api_key=GEMINI_API_KEY,
    model=GEMINI_MODEL,
    temperature=0.3,
)
```

```python
class CodeState(TypedDict):
    """코드 생성 에이전트의 상태 정의.

    필드:
        use_case: 해결할 코딩 문제 설명 (한국어/영어)
        goals: 코드가 만족해야 할 목표 문자열 리스트
        iteration: 현재 반복 횟수 (0부터 시작)
        max_iterations: 최대 반복 횟수
        code: 가장 최근에 생성된 코드 (헤더 없는 순수 코드)
        feedback: 가장 최근 코드에 대한 리뷰/피드백 텍스트
        goals_met: 목표 충족 여부
        final_code: 최종 결과 코드 (목표를 충족했다고 판단된 시점의 코드)
    """

    use_case: str
    goals: List[str]

    iteration: int
    max_iterations: int

    code: Optional[str]
    feedback: Optional[str]

    goals_met: bool
    final_code: Optional[str]
```

```python
def generate_code_node(state: CodeState) -> dict:
    """현재 상태를 바탕으로 Python 코드를 새로 생성하거나 개선하는 노드.

    - use_case / goals를 기반으로 기본 코드 생성
    - code, feedback이 있으면 이를 이용해 리팩터링/수정 유도
    - 결과를 state["code"]에 저장하고 iteration을 +1
    """

    goals_list = "\n".join(f"- {g.strip()}" for g in state["goals"])

    prompt_parts: List[str] = [
        "너는 AI 코딩 에이전트입니다. 다음 사용 사례에 따라 Python 코드를 작성하거나 개선하세요.",
        "",
        f"사용 사례:\n{state['use_case']}",
        "",
        "다음 목표를 모두 충족해야 합니다:",
        goals_list,
    ]

    if state.get("code"):
        prompt_parts.append("\n이전 코드:\n" + state["code"])

    if state.get("feedback"):
        prompt_parts.append("\n이전 코드에 대한 피드백:\n" + state["feedback"])

    prompt_parts.append(
        "\n수정된 파이썬 코드만 반환하세요. "
        "코드 외의 자연어 설명이나 주석(설명용 주석)은 포함하지 마세요."
    )

    prompt = "\n".join(prompt_parts)

    resp = llm.invoke([HumanMessage(content=prompt)])
    code = str(resp.content).strip()

    return {
        "code": code,
        "feedback": None,
        "iteration": state["iteration"] + 1,
    }
```

```python
def review_code_node(state: CodeState) -> dict:
    """현재 코드에 대해 목표 기준으로 리뷰/피드백을 생성하는 노드."""

    code = state.get("code")
    if not code:
        return {"feedback": None}

    goals_list = "\n".join(f"- {g.strip()}" for g in state["goals"])

    prompt = (
        "너는 Python 코드 리뷰어입니다. 아래 코드를 보고, 주어진 목표를 얼마나 만족하는지 평가하고 "
        "개선해야 할 부분을 구체적으로 제안하세요.\n\n"
        f"목표:\n{goals_list}\n\n"
        f"코드:\n{code}\n\n"
        "다음 관점에서 부족한 점이 있다면 모두 언급하세요:\n"
        "- 명확성\n"
        "- 단순성\n"
        "- 정확성\n"
        "- 엣지 케이스 처리\n"
        "- 테스트 가능성 또는 테스트 고려 여부\n"
    )

    resp = llm.invoke([HumanMessage(content=prompt)])
    feedback = str(resp.content).strip()

    return {"feedback": feedback}
```

```python
def evaluate_goals_node(state: CodeState) -> dict:
    """피드백을 기반으로 목표 달성 여부를 판단하는 노드.

    - 반복 횟수가 max_iterations에 도달하면 강제 종료(goals_met=True)
    - 피드백이 없으면 False
    - LLM에게 True/False만 답하게 하여 목표 달성 여부를 판단
    """

    if state["iteration"] >= state["max_iterations"]:
        return {"goals_met": True}

    feedback = state.get("feedback")
    if not feedback:
        return {"goals_met": False}

    goals_list = "\n".join(f"- {g.strip()}" for g in state["goals"])

    prompt = (
        "너는 AI 코드 품질 검토자입니다.\n\n"
        "아래 목표와 코드 리뷰 내용을 바탕으로, 현재 코드가 목표를 충분히 만족했다고 볼 수 있는지 판단하세요.\n\n"
        f"목표:\n{goals_list}\n\n"
        "코드 리뷰 내용:\n"
        f"{feedback}\n\n"
        "목표를 충분히 만족했다고 판단되면 True, 아직 부족하다고 판단되면 False 한 단어로만 답하세요."
    )

    resp = llm.invoke([HumanMessage(content=prompt)])
    answer = str(resp.content).strip().lower()

    goals_met = "true" in answer
    return {"goals_met": goals_met}
```

```python
def finalize_code_node(state: CodeState) -> dict:
    """최종 코드를 상태에 저장하는 노드.

    - 여기서는 파일 저장을 하지 않고, state["final_code"]에만 저장
    - 필요하다면 호출 측에서 final_code를 파일로 쓰면 됨
    """

    code = state.get("code")
    if not code:
        return {"final_code": None}

    final_code = code.strip() + "\n"
    return {"final_code": final_code}
```

```python
graph = StateGraph(CodeState)

graph.add_node("generate_code", generate_code_node)
graph.add_node("review_code", review_code_node)
graph.add_node("evaluate_goals", evaluate_goals_node)
graph.add_node("finalize_code", finalize_code_node)

graph.add_edge(START, "generate_code")
graph.add_edge("generate_code", "review_code")
graph.add_edge("review_code", "evaluate_goals")

def decide_next_node(state: CodeState) -> str:
    if state["goals_met"]:
        return "finalize_code"

    if state["iteration"] >= state["max_iterations"]:
        return "finalize_code"

    return "generate_code"

graph.add_conditional_edges("evaluate_goals", decide_next_node)
graph.add_edge("finalize_code", END)

checkpointer = InMemorySaver()
app = graph.compile(checkpointer=checkpointer)
```

```python
initial_state: CodeState = {
    "use_case": "주어진 양의 정수의 BinaryGap 길이를 구하는 코드를 작성하시오.",
    "goals": ["코드는 이해하기 쉬워야 한다", "정확해야 한다", "엣지 케이스를 처리해야 한다"],
    "iteration": 0,
    "max_iterations": 3,
    "code": None,
    "feedback": None,
    "goals_met": False,
    "final_code": None,
}

config = {
    "configurable": {
        "thread_id": "example_thread_001"
    }
}

final_state = app.invoke(initial_state, config=config)
```
